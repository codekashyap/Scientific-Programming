### Develop a python program that reads the image, display matrix representation of an image creates a histogram of the image and apply the smoothing effect on an image

```python
import numpy as np
from PIL import Image
from matplotlib import pyplot as plt
from skimage import io
import os
camera = io.imread('image.jpg')
print(camera.shape)
print(camera.ndim)
print(type(camera))
print(camera.dtype)
plt.imshow(camera,cmap=plt.cm.gray)
print(camera)
```

```python
ax = plt.hist(camera.ravel(), bins = 256)
plt.xlabel('Intensity Value')
plt.ylabel('Count')
plt.title('Intensity Value vs C')
plt.show()
```

```python
import scipy
from scipy import ndimage
lena = io.imread('image.jpg')
f = lena.astype(int)
blurred_f = ndimage.gaussian_filter(f, 1.5)
plt.figure(figsize=(12, 4))
plt.subplot(121)
plt.imshow(f, cmap=plt.cm.gray)
plt.axis('off')
plt.subplot(122)
plt.imshow(blurred_f, cmap=plt.cm.gray)
plt.axis('off')
plt.tight_layout()
plt.show()
```
