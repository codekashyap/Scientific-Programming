Pandas allows us to fill missing value using fillna() function. We can use mean(), median(), mode() and any self-defined function to calculate filling value.

### 4. Visualizing data

```python
plt.rcParams["figure.figsize"] = (16,8)
```

##### Part (a)

```python
a = df['Real coffee'].tolist()
b = df['Hot Tea'].tolist()
c = df['Country'].tolist()
df1 = pd.DataFrame({'Real coffee': a ,
                   'Hot Tea':b}, index = c)
df1.plot.bar(rot=0)
plt.title("Real coffee vs Hot Tea")
plt.xlabel("%")
plt.ylabel("Country")
plt.show()
```

#### Here, we analysis the bar plot between different contries and their population% who use real coffee and hot tea. It helps to make different conclusion about consumption of real coffee and hot tea.

##### Part (b)

```python
l= ['Real coffee','Instant coffee','Hot Tea','Sweetener','Biscuits','Powder soup','Tin soup','Potatoes',
    'Frozen fish','Frozen veggies','Apples','Oranges','Tinned fruit','Jam','Garlic','Butter','Margarine','Olive oil','Yoghurt','Crisp bread']
data=[]
for i in l:
    a=df.loc[df[i] == max(df[i])]['Country'].tolist()[0]
    b=df.loc[df[i] == max(df[i])][i].tolist()[0]
    data.append([a,b])
dfz = pd.DataFrame(data, columns=['Country', '%'], index=l)
```

```python
dfz['%'].plot.bar()
plt.title("Food and its maximum consuption by the share of population")
plt.xlabel("%")
plt.ylabel("Food")
plt.show()
```

#### Part (c)

```python
a = df['Apples'].tolist()
b = df['Oranges'].tolist()
c = df['Country'].tolist()
df1 = pd.DataFrame({'Apple': a ,
                   'Orange':b}, index = c)
df1.plot.bar(rot=0, color={'Apple': "red", 'Orange': "orange"})
plt.title("Apple vs Orange")
plt.xlabel("%")
plt.ylabel("Country")
plt.show()
```

#### Here, we analysis the bar plot between different contries and their population% who use apple and oranges. It helps to make different conclusion about consumption of apple and oranges.
