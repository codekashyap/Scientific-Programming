```python
import pandas as pd
```

```python
data1=pd.read_csv("user_device.csv")
```

```python
data1.head(3)
```

```python
data2=pd.read_csv("user_usage.csv")
```

```python
data2.head(3)
```

```python
merged_data = data1.merge(data2,on=["use_id"])
```

```python
merged_data
```

```python
df=merged_data.groupby(['device']).sum('outgoing_mins_per_month')
```

```python
df
```

```python
df.sample(3).plot(kind='pie', y='outgoing_mins_per_month', autopct='%1.0f%%')
```

```python
df=merged_data.groupby(['device']).sum('outgoing_sms_per_month').sample(3).plot(kind='pie', y='outgoing_mins_per_month', autopct='%1.0f%%')
```
