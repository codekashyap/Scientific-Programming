## Factorial of the given number

```python
def factorial(n):
    if n == 0:
        return 1
    else:
        return n*factorial(n-1)
```

```python
n=int(input("Enter the no: "))
factorial(n)
```

## First n Fibonacci numbers

```python
def fibo(n):
    a=0
    b=1
    count=2
    l=[0,1]
    while(count != n):
        c=a+b
        l.append(c)
        a=b
        b=c
        count+=1
    return l       
```

```python
n=int(input("Enter the no: "))
fibo(n)
```
